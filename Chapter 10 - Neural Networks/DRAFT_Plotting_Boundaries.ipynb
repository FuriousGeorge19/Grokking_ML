# Plotting the boundary of a continuous neural network

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import *
```

```python
alien_dataset = pd.DataFrame({
    'aack':[1,2,0,0,1,1,2,2],
    'beep':[0,0,1,2,1,2,1,2],
    'happy': [0,0,0,0,1,1,1,1]})

X = alien_dataset[['aack', 'beep']]
y = alien_dataset['happy']
```

### Plot obtained with the step activation function

```python
def step(x):
    if x >= 0:
        return 1
    else:
        return 0

def f(a,b):
    return step(6*a + 10*b - 15) + step(10*a + 6*b - 15) - 1.5
```

```python
xx, yy = np.meshgrid(np.arange(-0.5,3,0.005),
                     np.arange(-0.5,3,0.005))
Z = np.array([f(i[0], i[1]) for i in np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, colors=['red', 'blue'], alpha=0.25)#, levels=range(-1,2))
plt.contour(xx, yy, Z,colors = 'k',linewidths = 3)
plot_points(X, y, 100)
plt.show()
```

```python
def line_1(a,b):
    return step(6*a + 10*b - 15)

def line_2(a,b):
    return step(10*a + 6*b - 15)
```

```python
xx, yy = np.meshgrid(np.arange(-0.5,3,0.005),
                     np.arange(-0.5,3,0.005))
Z = np.array([line_1(i[0], i[1]) for i in np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, colors=['red', 'blue'], alpha=0.25, levels=range(-1,2))
plt.contour(xx, yy, Z,colors = 'k',linewidths = 3)
plot_points(X, y, 100)
plt.show()
```

```python
xx, yy = np.meshgrid(np.arange(-0.5,3,0.005),
                     np.arange(-0.5,3,0.005))
Z = np.array([line_2(i[0], i[1]) for i in np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, colors=['red', 'blue'], alpha=0.25, levels=range(-1,2))
plt.contour(xx, yy, Z,colors = 'k',linewidths = 3)
plot_points(X, y, 100)
plt.show()
```

### Plot obtained with the sigmoid activation function

```python
def sigmoid(x):
    return np.exp(x)/(1.0 + np.exp(x))

def g(a,b):
    return step(sigmoid(6*a + 10*b - 15) + sigmoid(10*a + 6*b - 15) - 1.5)
```

```python
xx, yy = np.meshgrid(np.arange(0,3,0.01),
                     np.arange(0,3,0.01))
Z = np.array([g(i[0], i[1]) for i in np.c_[xx.ravel(), yy.ravel()]])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, colors=['red', 'blue'], alpha=0.25, levels=[-0.5,0.5,1.5])
plt.contour(xx, yy, Z,colors = 'k',linewidths = 3)
plot_points(X, y, point_size=100)
plt.show()
```
