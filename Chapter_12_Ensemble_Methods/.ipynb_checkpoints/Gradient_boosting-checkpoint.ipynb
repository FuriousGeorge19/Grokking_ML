```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
import utils
import matplotlib.pyplot as plt
```

```python
def plot_regressor(model, features, labels):
    x = np.linspace(0,85,1000)
    plt.scatter(features, labels)
    plt.plot(x, model.predict(x.reshape([-1,1])))
    plt.xlabel("Age")
    plt.ylabel("Days per week")
    plt.show()
```

```python
features = np.array([[10],[20],[30],[40],[50],[60],[70],[80]])
labels = np.array([7,5,6,0,1,0,4,3])
```

```python
plt.scatter(features, labels)
plt.xlabel("Age")
plt.ylabel("Days per week")
plt.show()
```

```python
dt_regressor = DecisionTreeRegressor(max_depth=2)
dt_regressor.fit(features, labels)
```

```python
utils.display_tree(dt_regressor)
```

```python
plot_regressor(dt_regressor, features, labels)
```

# Gradient boosting

```python
residuals1 = labels - labels.mean()
residuals1
```

```python
gb1 = DecisionTreeRegressor(max_depth=2)
gb1.fit(features, residuals1)
```

```python
plot_regressor(gb1, features, residuals1)
```

```python
utils.display_tree(gb1)
```

```python
learning_rate = 0.9
gb = GradientBoostingRegressor(n_estimators=1, max_depth=2, alpha=learning_rate, verbose=10)
```

```python
gb.fit(features, labels)
```

```python
plot_regressor(gb, features, labels)
```

```python
labels = labels-labels.mean()
print(labels)
residuals = labels
for tree in gb.estimators_:
    print("******Tree******")
    predictions = tree[0].predict(features)
    print('predictions', predictions)
    residuals = residuals - predictions*learning_rate
    plot_regressor(tree[0], features, labels)
    print('residuals', residuals)
```

```python
residuals = [labels]
predictions = np.zeros(len(labels))
print(residuals)
print(predictions)
for i in range(len(gb.estimators_)):
    tree = gb.estimators_[i][0]
    current_predictions = tree.predict(features)
    predictions = predictions + learning_rate*current_predictions
    print(current_predictions)
```

```python
learning_rate=0.9
for i in range(1,100):
    gb = GradientBoostingRegressor(max_depth=2, n_estimators=i, learning_rate=learning_rate)
    gb.fit(features, labels)
    print(i)
    print(gb.predict(features))
```

```python
for i in range(len(gb.estimators_)):
    tree = gb.estimators_[i][0]
    current_predictions = tree.predict(features)
    print(current_predictions)
    plot_regressor(tree, features, predictions)
```

```python
alpha=0.9
labels
dt = DecisionTreeRegressor(max_depth=2)
dt.fit(features, labels)
plot_regressor(dt, features, labels)
predictions = dt.predict(features)
print(predictions)
residuals = labels-predictions
print(residuals)
dt2 = DecisionTreeRegressor(max_depth=2)
dt2.fit(features, residuals)
plot_regressor(dt2, features, residuals)
predictions2 = predictions + alpha*dt2.predict(features)
print(predictions2)
```
